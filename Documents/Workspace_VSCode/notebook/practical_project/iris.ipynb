The problems:

A customer hires you to give him/her an advise about choosing the appropriate Iris-class-flower based on 2 features about the size of petal and sepal.

Example: if he wants to have a class of Iris flowers which contains small sepal-size and large petal-size, which class will you advice him to choose?

```python
src = "C:/Users/DuyAnh/Documents/Workspace(No Coding)/Data Science/data_to_train/iris.csv"
```

```python
import pandas as pd
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(src, header=None)
```

```python
header = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']

df.columns = header
```

```python
df.info()
```

```python
df['class'].value_counts()
```

```python
setosa = df[df['class'] == 'Iris-setosa']
versicolor = df[df['class'] == 'Iris-versicolor']
virginica = df[df['class'] == 'Iris-virginica']
```

```python
setosa.describe()
```

```python
versicolor.describe()
```

```python
virginica.describe()
```

```python
setosa.drop(['class'], axis=1, inplace=True)
```

```python
setosa['sepal_area'] = list((np.array(setosa['sepal_length']) * np.array(setosa['sepal_width'])) / 4 * np.pi)
```

```python
setosa['sepal_area'] = round(setosa['sepal_area'], 2)
```

```python
setosa['l_over_w'] = list(np.array(setosa['sepal_length']) / np.array(setosa['sepal_width']))
setosa['l_over_w'] = round(setosa['l_over_w'], 2)
```

```python
setosa.drop(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], axis = 1, inplace=True)
```
